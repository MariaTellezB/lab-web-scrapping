**Instructions:**

````python
#your code

names = soup_td.find_all('h1',class_ ='h3')

dev_names = []

for n in names:
    n = str(n)
    n = re.sub(r'.*">', '', n)
    n = n.strip("</a></h1>, </a>, \n")
    dev_names.append((n))

print(dev_names)
````

````python
user = soup_td.find_all('p', class_='text-normal')

user_names = []

for u in user:
    u = str(u)
    u = re.sub(r'.*">', '', u)
    u = u.strip("\n</a>\n</p, </a>, \n")
    user_names.append((u))

user_names.insert(5,'')
print(user_names)
````

####  20 latest earthquakes info (date, time, latitude, longitude and region name) by the EMSC as a pandas dataframe

````python
# This is the url you will scrape in this exercise
url = 'https://www.emsc-csem.org/Earthquake/'
html = requests.get(url).content

soup_quakes = bs(html, 'lxml')
````

````python
quake_date_time = soup_quakes.find_all('b')[:20]
quake_dt = [str(d).split('>')[-3].split() for d in quake_date_time]

quake_date= [d[0] for d in quake_dt]
quake_time = [t[1].replace('</a','') for t in quake_dt]
````

````python
quake_latitude = soup_quakes.find_all('td',class_='tabev1')[:20]
quake_l = [str(l).split('>')[1].replace('</td','').split()[0] for l in quake_latitude]

quake_latitude_l = soup_quakes.find_all('td', class_='tabev2')[:20]
quake_ll= [str(l).split('>')[1].replace('</td','').split()[0] for l in quake_latitude_l]

quake_lat = [quake_l[i] +' '+ quake_ll[i] for i in range(len(quake_l))]
````

````python
quake_longitude = soup_quakes.find_all('td',class_='tabev1')[1::2][:20]
quake_lo = [str(l).split('>')[1].replace('</td','').split()[0] for l in quake_longitude]

quake_lon_l = soup_quakes.find_all('td', class_='tabev2')[1::3]
quake_llon= [str(l).split('>')[1].replace('</td','').split()[0] for l in quake_lon_l]

quake_lon = [quake_lo[i] +' '+ quake_llon[i] for i in range(len(quake_lo))]                                         
````

````python
quake_region = soup_quakes.find_all('td', class_='tb_region')[:20]
quake_reg = [str(r).split('>')[-2].replace('\xa0','').replace('</td','') for r in quake_region]

quake_reg
````

````python
earthqukes = {'date': quake_date, 
          'time': quake_time, 
          'latitude': quake_lat, 
          'longitude': quake_lon, 
          'region': quake_reg}
````

````python
earthquakes_df = pd.DataFrame(earthqukes)

earthquakes_df
````

#### Display the date, days, title, city, country of next 25 hackathon events as a Pandas dataframe table

````python
# This is the url you will scrape in this exercise
url ='https://mlh.io/seasons/na-2020/events'
html = requests.get(url).content

soup_hack = bs(html, 'lxml')
````

````python
t = soup_hack.find_all('h3',class_='event-name')[:26]
hack_title = [str(i).split('>')[1].replace('</h3','') for i in t]
````

````python
h_d = soup_hack.find_all('p', class_='event-date')[:26]
hack_date = [str(i).split('>')[1].replace('       </p','') for i in h_d]
````

````python
h_c = soup_hack.select('div.event-location > span[itemprop="city"]')[:26]
hack_city = [str(i).split('>')[1].replace('</span','') for i in h_c]

h_s = soup_hack.select('div.event-location > span[itemprop="state"]')[:26]
hack_state = [str(i).split('>')[1].replace('</span','') for i in h_s]

hack_loc = [hack_city[i] + ',' + hack_state[i]  for i in range(len(hack_city))]
````

hack_d = {'date': hack_date, 
          'title': hack_title, 
          'location': hack_loc}

hack_df = pd.DataFrame(hack_d)

#### Top 10 languages by number of native speakers stored in a Pandas Dataframe

````python
# This is the url you will scrape in this exercise
url = 'https://en.wikipedia.org/wiki/List_of_languages_by_number_of_native_speakers'
html = requests.get(url).content

soup_nlan = bs(html, 'lxml')
````

````python
lan = soup_nlan.select('tbody > tr > td')[1::6][0:10]

t_lan =[i.text.strip('href').strip('\n') for i in soup_nlan.select('tbody > tr > td')[1::6][0:10]]
t_lan
````

````python
t_nat =[i.text.strip('href').strip('\n') for i in soup_nlan.select('tbody > tr > td')[2::6][0:10]]
t_nat
````

````python
top_lan = {'Language': t_lan, 
          'Native Spealers (millions)': t_nat}

top_lan_df = pd.DataFrame(top_lan)
top_lan_df
````
